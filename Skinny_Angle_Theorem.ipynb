```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Setup the range for theta (in radians)
theta = np.linspace(0, 1.5, 500)
g_theta = np.sin(theta)
g_prime_theta = theta

# 2. Calculate the percent error
# Error = |(approx - actual) / actual|
error = np.abs((g_prime_theta - g_theta) / g_theta)

# 3. Find the index where error first exceeds 5%
threshold = 0.05
indices_within_5pct = np.where(error <= threshold)[0]
max_theta = theta[indices_within_5pct[-1]]
max_degrees = np.degrees(max_theta)

# 4. Plotting
plt.figure(figsize=(10, 6))
plt.plot(theta, g_theta, label=r'$g(\theta) = \sin \theta$', linewidth=2)
plt.plot(theta, g_prime_theta, '--', label=r"$g'(\theta) = \theta$", color='orange')

# Mark the 5% threshold
plt.axvline(max_theta, color='red', linestyle=':', label=f'5% Error Boundary ({max_theta:.3f} rad)')

plt.title('Small-Angle Approximation vs. Sine Function')
plt.xlabel('Theta (radians)')
plt.ylabel('Value')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"The approximation is accurate within 5% up to θ ≈ {max_theta:.3f} radians ({max_degrees:.1f}°).")
```
