```python
import numpy as np
import matplotlib.pyplot as plt

h = 6.626e-34
c = 2.998e8
kB = 1.381e-23

def planck_law(wavelength, T):
  a = 2.0 * h * c**2
  b = (h * c) / (wavelength * kB * T)
  intensity = a / (wavelength**5 * (np.exp(b) - 1.0))
  return intensity

wavelengths = np.linspace(100e-9, 2500e-9, 1000)
wavelengths_angstroms = wavelengths * 1e10

temperatures = [3000, 4000, 5000, 6000]

plt.figure(figsize=(10, 6))

for T in temperatures:
  intensity = planck_law(wavelengths, T)
  plt.plot(wavelengths_angstroms, intensity, label=f'{T} K')

plt.axvspan(4000, 7000, color='gray', alpha=0.2, label='Visible Light')

plt.title("Blackbody Spectra (Planck Curves)")
plt.xlabel("Wavelength ($\mathring{A}$)")
plt.ylabel("Intensity ($B_\lambda$)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(0, 25000)
plt.tight_layout()

plt.show()
```
