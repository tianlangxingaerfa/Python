```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 500)

frequencies = [1, 2, 5]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

plt.figure(figsize=(10, 6))

for w, color in zip(frequencies, colors):
  y = np.sin(2 * np.pi * w * t)
  plt.plot(t, y, label=f' = {w}', color=color, linewidth=2)

  peak_t = 1 / (4 * w) + 1 / w
  plt.text(peak_t, 1.1, f' = {w}', color=color,
             ha='center', fontweight='bold')

plt.title(r'Plot of $\sin(2\pi\omega t)$ for three various frequencies', fontsize=14)
plt.xlabel('Time(t)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.ylim(-1.5, 1.5)
plt.grid('True', linestyle='--', alpha=0.7)
plt.axhline(0, color='black', linewidth=1)

plt.show()
```
